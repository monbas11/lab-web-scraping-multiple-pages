```python
def get_songs(soup):
    
    song = [h4.get_text().split(" - ") for h4 in soup.find_all('h4') if h4.get_text(strip=True) not in [" Top Charts", " Categories / Genres", "\nNavigation\n","\n Stay in touch\n", "\n Our office\n", "About us", " Countries"]]

    return song
```

```python
top_songs =[]

for i in range (0, 5):
    c= i*20
    r = requests.get(f"https://www.icharts.co.za/us/music/topsongs/21-rock/?start={c}&length=20/", headers=header)
    soup = BeautifulSoup(r.content, 'html.parser')
    top_songs.extend(get_songs(soup))
    
    
    
pd.DataFrame(top_songs[3:-4:2], columns=["song", "artist"])
```

```python
data = pd.DataFrame(top_songs[3:-4:2], columns=["song", "artist"])
```
